```python
#data distribution and visualization

#set up matplotlib figure
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=[15,15])

data['Gender'].value_counts().plot(kind='barh', ax=axes[0,0], title='Gender', color=['blue', 'gray'])
data['Age'].value_counts().plot(kind='barh', ax=axes[0,1], title='Age')
data['City_Category'].value_counts().plot(kind='barh', ax=axes[1,0], title='City_Category', color=['gray', 'black', 'teal'])
data['Marital_Status'].value_counts().plot(kind='barh', ax=axes[1,1], title='Marital_Status')
data['Occupation'].value_counts().plot(kind='barh', ax=axes[2,0], title='Occupation')
data['Stay_In_Current_City_Years'].value_counts().plot(kind='barh', ax=axes[2,1], title='Stay_In_Current_City_Years')
```

```python
# from above: much more data about male consumer than female and some occupations represented more than others
# Also, call-out for the stay_in_current_city classifier- most only stay for 1 year! implications- independent, not ready to settle, young career?
# city B, 0 (not married), occupation 4 
```

```python
# heatmap 1

# makes the matplotlib figure
f, ax = plt.subplots(figsize=(12, 12))

#heatmap command
sns.heatmap(data.corr(),annot = True, linewidths=.1, fmt= '.2f',mask=np.zeros_like(data.corr(), dtype=np.bool), cmap=sns.diverging_palette(620, 650, as_cmap=True), square=True, ax=ax)
plt.title('Black Friday Heatmap')
plt.show()
# this heatmap visually displays correlation, which shows the relationship numerically
# see that correlation between product_category_2 and product_category_3 are the largest 
```
